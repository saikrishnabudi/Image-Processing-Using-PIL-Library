# **Converting the image into Gray Scale**

```python
# Load the image and saving it into Gray Scale
import IPython.display as display
from PIL import Image

# Loading the image
image=Image.open("/content/bille.jpg")

# Converting the image into Gray Scale
gs_image=image.convert(mode="L")              # Mode = "L" is used to convert the image into gray scale

# Saving the image into JPEG format
gs_image.save("/content/bille.jpg")

# Loading the image and showing it
image2=Image.open("/content/bille.jpg")

# Displaying the Image
display.display(image2)
```

# **Re-Sizing the Image**

```python
import IPython.display as display
from PIL import Image

# Loading the image
image=Image.open("/content/bille.jpg")

# Sizing the image
print(image.size)

# Creating the thumbnail and preserving the aspect ratio
image.thumbnail((100,100))

# Reporting the size & modifying the size
print(image.size)

# Dsiplaying the image results
image.show()
display.display(image)
```

- It resize the image and maintain the aspect ratio.

# **Flipping, Rotating, Cropping the Image**

```python
from PIL import Image
from matplotlib import pyplot

# Load Image
image = Image.open("/content/bille.jpg")

# Horizontal Flip
horz_flip = image.transpose(Image.FLIP_LEFT_RIGHT)

# Vertical Flip
vert_flip = image.transpose(Image.FLIP_TOP_BOTTOM)

# Plotting all the three images with increased visualization size
pyplot.figure(figsize=(10, 6))  # Set the figure size

# Plotting Original Image
pyplot.subplot(1, 3, 1)  # Subplot with 1 row, 3 columns, and index 1
pyplot.imshow(image)
pyplot.title('Original')

# Plotting for Horizontal Flip
pyplot.subplot(1, 3, 2)  # Subplot with 1 row, 3 columns, and index 2
pyplot.imshow(horz_flip)
pyplot.title('Horizontal Flip')

# Plotting for Vertical Flip
pyplot.subplot(1, 3, 3)  # Subplot with 1 row, 3 columns, and index 3
pyplot.imshow(vert_flip)
pyplot.title('Vertical Flip')

# Displying the results
pyplot.show()
```
